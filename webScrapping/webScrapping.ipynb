```
import requests
```

## Basic request
---

```
response = requests.get('https://platzi.com')
print(dir(response))
```

```
print(f'Status code: {response.status_code}')
print(f'Headers:\n\tRequest date: {response.headers["date"]}')
```

```
print('Text'.center(30))
print('-' * 30)
print(response.text)
```

## Extract info from HTML document
---

```
import bs4
```

```
soup = bs4.BeautifulSoup(response.text, 'html.parser')

courses_urls = []
for course in soup.select('.RecentCourses-item > a'):
  courses_urls.append(course['href'])

for course_url in courses_urls:
  print(course_url[8:])
```

## Getting info from a news website
---

```
response_news = requests.get('https://www.eluniversal.com.mx/')
html_news = bs4.BeautifulSoup(response_news.text, 'html.parser')
```

```
news_links = [news_item['href'] for news_item in html_news.select('.field-content > a')]

for i, link in enumerate(news_links):
  if link[0] == '/':
    news_links[i] = 'https://www.eluniversal.com.mx' + link

print(news_links)
```
